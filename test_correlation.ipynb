```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
sample_rate = 20
length_sec = 1000
length = sample_rate * length_sec
time = np.linspace(0, length_sec, length)
freqs = [1/30, 1/40, 1/60, 1/80]
signal_one = np.zeros(length)
signal_two = np.zeros(length)
for freq in freqs:
    signal_one += np.sin(2 * np.pi * freq * time + np.random.rand() * np.pi/8)
    signal_two += np.sin(2 * np.pi * freq * time + np.random.rand() * np.pi/8)
# signal_one += np.random.normal(0, 1, length)
# signal_two += np.random.normal(0, 1, length)
plt.plot(time, signal_one)
plt.plot(time, signal_two)
plt.show()
print(pearsonr(signal_one, signal_two))
corr_win_list = []
sig_one_cat = []
sig_two_cat = []
window_size = 400
# create list of 5 numbers from 0 to 20
random_numbers = (np.random.rand(7) * 20).astype(int)
random_numbers.sort()
# remove all repeated indices
random_numbers = np.unique(random_numbers)
print(random_numbers)
for j in (random_numbers):
    sig_one_cat.extend(signal_one[j*window_size:(j+1)*window_size])
    sig_two_cat.extend(signal_two[j*window_size:(j+1)*window_size])
    corr_win_list.append(pearsonr(signal_one[j*window_size:(j+1)*window_size], signal_two[j*window_size:(j+1)*window_size])[0])
print(np.mean(corr_win_list))
print(pearsonr(sig_one_cat, sig_two_cat)[0])
```
