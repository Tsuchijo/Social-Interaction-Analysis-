```python
import sklearn as sk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
length = 10000
freq = 1000
x_1 = np.random.normal(0, 1, length) + np.sin(np.linspace(0, 2 * np.pi, length))
x_2 = np.random.normal(0, 1, length) + np.sin(np.linspace(0, 2 * np.pi, length)) + 10
```

```python
plt.plot(x_1)
plt.plot(x_2)
```

```python
# pearson correlation over whole
print('pearson correlation over whole')
print(stats.pearsonr(x_1, x_2)[0])
corr_running = []
iter = 10
window_length = length // iter
for i in range(iter):
    corr_running.append(stats.pearsonr(x_1[i * window_length : (i + 1) * window_length], x_2[i * window_length:(i + 1) * window_length])[0])
mean_corr = np.mean(corr_running)
print('mean correlation over windows')
print(mean_corr)
```
